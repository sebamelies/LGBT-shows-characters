```python
import pandas as pd

# Ruta al archivo Excel
archivo_excel = "/Users/sebasastre/Downloads/LGBT characters/20220601-LGBTQ-cartoons-series-information-v3.xlsx"

# Leer el archivo Excel
datos_excel = pd.read_excel(archivo_excel, header=1)

# Especifica la ruta del archivo CSV
file_path = '/Users/sebasastre/Downloads/LGBT characters/20220601-LGBTQ-character-demographics-v3.csv'

# Lee el archivo CSV y carga su contenido en un DataFrame de Pandas
df = pd.read_csv(file_path)

# Fusionar los DataFrames
merged_df = pd.merge(datos_excel, df, on='ID', how='inner')

# Mostrar el DataFrame fusionado
merged_df
```

```python
# Contar el número de filas con el valor "Cis Man" en la columna "gender"
count_cis_man = df['gender'].value_counts()['Cis Man']

print("Número de filas con 'Cis Man' en la columna 'gender':", count_cis_man)
```

```python
# Contar la cantidad de cada valor único en la columna "gender"
gender_counts = df['gender'].value_counts()

print("Cantidad de cada valor único en la columna 'gender':")
print(gender_counts)
```

```python
# Obtener todas las categorías distintas de la columna "orientation" y contar la cantidad de cada una
orientation_counts = df['orientation'].value_counts()

# Imprimir las categorías únicas en la columna 'orientation' junto con su cantidad
print("Categorías únicas en la columna 'orientation' y su cantidad:")
print(orientation_counts)
```

```python
# Agrupar por la columna "network" y contar la cantidad de personajes únicos en cada una
network_characters_counts = merged_df.groupby('network')['character_name'].nunique()

# Ordenar los resultados en orden descendente según la cantidad de personajes
network_characters_counts_sorted = network_characters_counts.sort_values(ascending=False)

# Mostrar las redes distintas y la cantidad de personajes únicos en cada una, ordenadas de mayor a menor
print("Redes distintas y cantidad de personajes únicos en cada una, ordenadas de mayor a menor:")
print(network_characters_counts_sorted)
```

```python
# Primero, eliminamos duplicados de personajes en diferentes shows
unique_characters = merged_df.drop_duplicates(subset=['character_name', 'show_title_y'])

# Luego, agrupamos por 'show_title' y contamos la cantidad de personajes únicos
show_characters_counts = unique_characters.groupby('show_title_y').size()

# Ordenamos los resultados en orden descendente según la cantidad de personajes únicos
show_characters_counts_sorted = show_characters_counts.sort_values(ascending=False)

# Seleccionamos los primeros 20 resultados
top_20_show_titles = show_characters_counts_sorted.head(20)

# Mostramos los resultados
print("Los 20 show_title con la mayor cantidad de character_name indistintos:")
print(top_20_show_titles)
```

```python
# Agrupar por la columna "network" y contar la cantidad de títulos de espectáculos únicos
show_titles_by_network = merged_df.groupby('network')['show_title_x'].nunique()

# Mostrar los resultados
print("Cantidad de títulos de shows únicos por Network:")
print(show_titles_by_network)
```

```python
# Filtrar el DataFrame combinado por el título del espectáculo
she_ra_df = merged_df[merged_df['show_title_x'] == 'She-Ra and the Princesses of Power']

# Seleccionar las columnas relevantes
she_ra_characters_info = she_ra_df[['character_name', 'race', 'orientation', 'gender']]

# Mostrar los resultados
print("Información de personajes de She-Ra and the Princesses of Power:")
print(she_ra_characters_info)
```

```python
# Filtrar el DataFrame combinado por el rol "Main Character"
main_characters_df = merged_df[merged_df['role'] == 'Main Character']

# Mostrar los resultados
print("Personajes cuyo rol es Main Character:")
main_characters_df
```

```python
# Filtrar el DataFrame combinado por el rol "Main Character"
main_characters_df = merged_df[merged_df['role'] == 'Main Character']

# Obtener una lista de los personajes únicos cuyo rol es "Main Character"
main_characters_list = main_characters_df['character_name'].unique()

# Mostrar la lista de personajes
print("Personajes cuyo rol es Main Character:")
print(main_characters_list)
```

```python
# Contar la cantidad de personajes cuyo rol es "Main Character"
cantidad_main_characters = len(main_characters_list)

# Mostrar la cantidad de personajes
print("Cantidad de personajes cuyo rol es Main Character:", cantidad_main_characters)
```

```python
# Contar la cantidad de shows distintos en la columna 'show_title_x'
cantidad_de_shows = merged_df['show_title_x'].nunique()

# Mostrar la cantidad de shows distintos
print("Cantidad de shows distintos:", cantidad_de_shows)
```

```python
# Obtener las categorías únicas de la columna 'role'
categorias_role = merged_df['role'].unique()

# Mostrar las categorías únicas
print("Categorías de la columna 'role':")
print(categorias_role)
```

```python
# Contar la cantidad de cada categoría en la columna 'role'
role_counts = merged_df['role'].value_counts()

# Mostrar los resultados
print("Cantidad de cada categoría en la columna 'role':")
print(role_counts)
```

```python
# Counting the number of unique character names
character_count = merged_df['character_name'].nunique()

# Printing the count of unique character names
print("Cantidad de nombres de personajes únicos:", character_count)
```

```python
# Convert 'start_date' column to datetime
merged_df['start_date'] = pd.to_datetime(merged_df['start_date'])

# Extract year from 'start_date' column
merged_df['year'] = merged_df['start_date'].dt.year

# Count the number of shows per year
shows_per_year = merged_df.groupby('year')['show_title_x'].nunique()

# Print the result
print("Cantidad de shows por año:")
print(shows_per_year)
```

```python
# Cuenta la cantidad de shows por cada categoría de rating
rating_counts = merged_df['TV_rating'].value_counts()

# Imprime los resultados
print("Cantidad de shows por categoría de rating:")
rating_counts
```
